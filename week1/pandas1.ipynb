# about data
hotel: Jenis hotel (misalnya, Hotel Resor atau Hotel Kota).

is_canceled: Nilai 1 jika pemesanan dibatalkan, dan 0 jika pemesanan tidak dibatalkan.

lead_time: Jumlah hari antara tanggal masuk pesanan dan tanggal kedatangan.

arrival_date_year: Tahun kedatangan.

arrival_date_month: Bulan kedatangan.

arrival_date_week_number: Nomor minggu kedatangan.

arrival_date_day_of_month: Hari kedatangan dalam sebulan.

stays_in_weekend_nights: Jumlah malam akhir pekan (Sabtu atau Minggu) yang dipesan.

stays_in_week_nights: Jumlah malam hari kerja (Senin hingga Jumat) yang dipesan.

adults: Jumlah orang dewasa.

children: Jumlah anak-anak.

babies: Jumlah bayi.

meal: Jenis paket makanan yang dipesan.

country: Negara asal pelanggan.

market_segment: Penunjukan segmen pasar.

distribution_channel: Saluran distribusi pemesanan.

is_repeated_guest: Nilai 1 jika nama pemesanan adalah tamu yang berulang, dan 0 sebaliknya.

previous_cancellations: Jumlah pembatalan sebelumnya oleh pelanggan ini.

previous_bookings_not_canceled: Jumlah pemesanan sebelumnya yang tidak dibatalkan oleh pelanggan ini.

reserved_room_type: Jenis kamar yang dipesan.

assigned_room_type: Jenis kamar yang ditetapkan.

booking_changes: Jumlah perubahan yang dilakukan pada pemesanan dari waktu masuk hingga waktu check-in atau pembatalan.

deposit_type: Indikasi apakah pelanggan membayar uang muka, tidak ada uang muka, atau uang muka tidak dapat dikembalikan.

agent: ID agen perjalanan.

company: ID perusahaan/entitas yang melakukan pemesanan.

days_in_waiting_list: Jumlah hari pemesanan berada dalam daftar tunggu sebelum dikonfirmasi ke pelanggan.

customer_type: Jenis pelanggan, menunjukkan apakah pemesanan bersifat ad-hoc, grup, kontrak, atau transien.

adr: Tarif Harian Rata-Rata (Average Daily Rate), didefinisikan sebagai total semua transaksi pemesanan dibagi dengan jumlah malam menginap.

required_car_parking_spaces: Jumlah tempat parkir yang dibutuhkan pelanggan.

total_of_special_requests: Jumlah permintaan khusus yang dibuat pelanggan (misalnya, tempat tidur tambahan, lantai tinggi, dll.).

reservation_status: Status reservasi:

Canceled: Pemesanan dibatalkan oleh pelanggan.

Check-Out: Pelanggan telah check-out.

No-Show: Pelanggan tidak muncul dan tidak memberi tahu hotel.

reservation_status_date: Tanggal saat status terakhir diperbarui.
